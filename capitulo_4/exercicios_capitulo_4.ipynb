**Mas qual seria a diferença da função de polyline para a polygon? e por quê?**
Em polyline é necessário passar o parâmetro com o valor do ângulo (angle), o que requer que o usuário calcule manualmente os ângulos internos, que é mais trabalhoso. Então, para facilitar o processo, deixamos encapsulado as ações repetitivas de desenhar lados e virar a tartaruga em polyline e, assim, nós a utilizamos em polygon para reutilização de código - dividindo um problema maior em subproblemas, pois em polygon o valor do ângulo é calculado automaticamente baseado no número de segmentos que é passado.

**MAS O QUE SERIA UM ARCO?**
Um arco é uma parte de um círculo. É um segmento da circunferência do círculo, compreendido entre dois pontos distintos na borda do círculo.


- OBSERVAÇÃO: o uso inadequado de docstrings é para uma melhor compreensão do que o código faz, com relação aos seus cálculos também. Por questão de estudos, não há problema.

```python
'''
t = objeto turtle
r = raio do arco
angle = ângulo do arco em graus
'''

def arc(t, r, angle):
  arc_length = 2 * math.pi * r * angle / 360
  '''Quando desenhamos um arco, ele é apenas uma parte da circunferência completa.
  A fração da circunferência que o arco representa é determinada pelo ângulo central
  do arco em relação ao círculo completo (360 graus).'''

  n = int(arc_length / 3) + 1 # número de segmentos
  '''Calcula o número de segmentos para desenhar o arco. Divide o comprimento do
    arco por 3 para obter um número razoável de passos e adiciona 1 para garantir
    pelo menos um segmento.
  '''
  step_length = arc_length / n
  '''Calcula o comprimento de cada passo (segmento) dividindo o comprimento
  total do arco pelo número de segmentos.
  '''

  step_angle = float(angle) / n
  '''Calcula o ângulo que a tartaruga deve virar após desenhar cada segmento, dividindo
  o ângulo total pelo número de segmentos.
  '''

  polyline(t, n, step_length, step_angle)
  '''Usa a função polyline para desenhar o arco. '''
```

**BASE DA FLOR:**

```python
def arc(t, r, angle):
  arc_length = 2 * math.pi * r * angle / 360  # uma parte do circulo
  n = int(arc_length / 3) + 1                 # número de segmentos
  step_length = arc_length / n                # comprimento de cada segmento
  step_angle = float(angle) / n               # angulo de cada segmento

  polyline(t, n, step_length, step_angle)     # desenha o arco
```

```python
# definição da função que vai desenhar a pétala inteira

def petal(t, r, angle):
    # primeiro arco da pétala
    arc(t, r, angle) # desenha o arco
    tt.lt(180 - angle) # A fórmula 180 - angle é usada para ajustar a direção da tartaruga para garantir que os dois arcos se conectem de forma fluida

    # segundo arco da pétala (que forma a pétala)
    arc(t, r, angle) # desenha o arco
    tt.lt(180 - angle) # ajusta a direção da tartaruga
```

```python
petala1 = tt.initializeTurtle()
petal(petala1, 100, 60)
```

**FLOR 1:**

```python
# definição de função flower, que vai desenhar a flor completa

'''
t: o objeto turtle
n: o número de pétalas que a flor vai ter
r: o raio dos arcos das pétalas
angle: O ângulo dos arcos das pétalas
'''

def flower(t, n, r, angle):

    for i in range(n): # repete o código dentro do loop n vezes, onde n é o número de pétalas que a flor vai ter
        petal(t, r, angle) # desenha uma pétala completa (2 segmentos)

        '''
        Gira a tartaruga para a esquerda (lt é para "left") em um ângulo de
        360/n graus. Esse ângulo é o necessário para a tartaruga se posicionar
        para desenhar a próxima pétala, distribuindo uniformemente as pétalas ao redor do círculo.
        '''
        tt.lt(360/n)
```

```python
flor1 = tt.initializeTurtle()
flower(flor1, 7, 100, 60)
```

**FLOR 2**

```python
# definição de função flower, que vai desenhar a flor completa

'''
t: o objeto turtle
n: o número de pétalas que a flor vai ter
r: o raio dos arcos das pétalas
angle: O ângulo dos arcos das pétalas
'''

def flower(t, n, r, angle):

    for i in range(n): # repete o código dentro do loop n vezes, onde n é o número de pétalas que a flor vai ter
        petal(t, r, angle) # desenha uma pétala completa (2 segmentos)
        '''
        Gira a tartaruga para a esquerda (lt é para "left") em um ângulo de
        360/n graus. Esse ângulo é o necessário para a tartaruga se posicionar
        para desenhar a próxima pétala, distribuindo uniformemente as pétalas ao redor do círculo.
        '''
        tt.lt(360/n)

    # utilizando a recursividade,  esta chamada de função dentro dela mesma desenha uma flor menor no centro da flor maior, se ainda há pétalas a desenhar
    if n > 1:
        flower(t, n // 2, r / 2, angle)
```

```python
flor2 = tt.initializeTurtle()

flower(flor2, 10, 100, 60)  # Flor com 10 pétalas
```
