A função integrada **open** recebe o nome do arquivo como
um parâmetro e retorna um objeto de arquivo que você pode usar para ler o arquivo.

```python
fin = open('words.txt')
```

**fin** é um nome comum de objeto de arquivo usado para entrada de dados. O objeto de arquivo oferece vários métodos de leitura, inclusive readline, que lê caracteres no arquivo até chegar a um comando de nova linha, devolvendo o resultado como uma string:


```python
fin.readline()
```

A primeira palavra nesta lista específica é “aa”, uma espécie de lava. A sequência '\r\n' representa dois caracteres que representam espaços em branco (whitespace), um retorno de carro e uma nova linha, que separa esta palavra da seguinte.
O objeto de arquivo grava a posição em que está no arquivo, então se você chamar
readline mais uma vez, receberá a seguinte palavra:

```python
fin.readline()
```

A palavra seguinte é “aah”, uma palavra perfeitamente legítima, então pare de olhar para mim desse jeito. Ou, se é o whitespace que está incomodando você, podemos nos livrar dele com o método de string strip:


```python
line = fin.readline()
word = line.strip()
```

```python
word
```

Você também pode usar um objeto de arquivo como parte de um loop for. Este programa lê words.txt e imprime cada palavra, uma por linha:


```python
fin = open('words.txt')

for line in fin:
  word = line.strip()
  print(word)
```

Altere seu programa na seção anterior para imprimir apenas as palavras que não têm “e” e calcule a porcentagem de palavras na lista que não têm “e”.

```python
fin = open('words.txt')
cnt_palavras = 0
cnt_palavras_sem_e = 0

for line in fin:
  word = line.strip()

  cnt_palavras = cnt_palavras + 1

  if 'e' not in word:
    print(word)
    cnt_palavras_sem_e = cnt_palavras_sem_e + 1

pct_palavras_sem_e = (cnt_palavras_sem_e/cnt_palavras) * 100

print('\n')
print('São ' + str(round(pct_palavras_sem_e,2)) + '% de palavras sem a letra "e" ')
```

```python
def avoids():
  cnt_letras_proibidas = 0

  palavra = str(input("Digite uma palavra: "))

  for x in range(5):
    letras_proibidas = str(input('Digite uma letra que não podem conter na palavra: '))
    letras_proibidas += letras_proibidas

  for letra in letras_proibidas:
    if letra in palavra:
      cnt_letras_proibidas = cnt_letras_proibidas + 1

  print('\n')

  if cnt_letras_proibidas == 0:
    return True
  else:
    return False
```

<h2> Exercício 9.4 </h2>

Escreva uma função chamada uses_only que receba uma palavra e uma série de letras e retorne True, se a palavra só contiver letras da lista. Você pode fazer uma frase usando só as letras acefhlo? Que não seja “Hoe alfalfa?”

```python
def uses_only():
  serie_de_letras = ''
  cnt = 0

  # digitação da palavra
  while True:
    palavra = str(input('Digite uma palavra: ')).strip()

    # caso-base
    if palavra == '' or len(palavra) == 1:
      print('Não é possível digitar apenas uma letra!\n')
      continue
    else:
      break

  print('Agora vamos digitar as letras!\n')

  # digitação da série de letras
  while True:
    letra = str(input('Digite uma letra: ')).strip()

    if len(letra) > 1:
      print('Digite apenas uma letra por vez!\n')
      continue
    else:
       if cnt < 4:
          serie_de_letras = serie_de_letras + letra
          cnt = cnt + 1
       # caso-base
       elif cnt == 4:
          break

  cnt = 1

  # verificação se a palavra só contém as letras da lista
  for l in palavra:
    if l in serie_de_letras:
      cnt = cnt + 1

  print('\n')

  if cnt == len(palavra):
    return True
  else:
    return False
```
